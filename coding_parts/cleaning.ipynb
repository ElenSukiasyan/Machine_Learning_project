```python
# Adding a new feature - Age Group
new_data['AgeGroup'] = pd.cut(new_data['Age'], bins=[0, 25, 40, 100], labels=['Young', 'Adult', 'Senior'])

# Adding a new feature - Employer Size Category
def categorize_employer_size(size):
    if size in ['1-5', '6-25']:
        return 'Small'
    elif size in ['26-100', '100-500']:
        return 'Medium'
    else:
        return 'Large'

new_data['EmployerSizeCategory'] = new_data['NumEmployees'].apply(categorize_employer_size)

# Adding a new feature - Geographical Region
def get_geographical_region(country, state):
    if country == 'United States of America':
        return {
            'Northeast': ['Connecticut', 'Maine', 'Massachusetts', 'New Hampshire', 'Rhode Island', 'Vermont', 'New York', 'Pennsylvania', 'New Jersey'],
            'Midwest': ['Illinois', 'Indiana', 'Michigan', 'Ohio', 'Wisconsin', 'Iowa', 'Kansas', 'Minnesota', 'Missouri', 'Nebraska', 'North Dakota', 'South Dakota'],
            'South': ['Delaware', 'Florida', 'Georgia', 'Maryland', 'North Carolina', 'South Carolina', 'Virginia', 'District of Columbia', 'West Virginia', 'Alabama', 'Kentucky', 'Mississippi', 'Tennessee', 'Arkansas', 'Louisiana', 'Oklahoma', 'Texas'],
            'West': ['Arizona', 'Colorado', 'Idaho', 'Montana', 'Nevada', 'New Mexico', 'Utah', 'Wyoming', 'Alaska', 'California', 'Hawaii', 'Oregon', 'Washington']
        }.get(state, 'Other')
    else:
        return 'Non-US'

new_data['RemoteWorkPreference'] = new_data['WorkRemotely'].apply(lambda x: 'PreferRemote' if x == 'Always' else ('PreferOffice' if x == 'Never' else 'Neutral'))

print(new_data)
```

```python
# Adding several new features with higher range (not from 0 but from the middle (5, 6, 7))

import numpy as np

mental_health_features = [
    'GoodRelationships',
    'HelpOthers',
    'DecisionMaking',
    'CareForYourself',
    'EmotionalResilience'
]

for feature in mental_health_features:
    if feature == 'GoodRelationships':
        new_data[feature] = np.random.randint(5, 9, size=len(data))
    elif feature == 'HelpOthers':
        new_data[feature] = np.random.randint(4, 8, size=len(data))
    elif feature == 'DecisionMaking':
        new_data[feature] = np.random.randint(7, 10, size=len(data))
    elif feature == 'CareForYourself':
        new_data[feature] = np.random.randint(6, 10, size=len(data))
    elif feature == 'EmotionalResilience':
        new_data[feature] = np.random.randint(5, 9, size=len(data))

new_data[mental_health_features] = new_data[mental_health_features].clip(0, 10)

print(new_data)
```

```python
# Handling with ranges in NumEmployees column

def clean_and_replace(data, column_name):
    def process_value(value):
        if not isinstance(value, str):
            value = str(value)
        if not value.isdigit():
            if '-' in value:
                start, end = map(int, value.split('-'))
                return str((start + end) // 2)
            return ''.join(filter(str.isdigit, value))
        return value
    data[column_name] = data[column_name].apply(process_value)

    data[column_name].replace('', '0', inplace=True)

# Activating
clean_and_replace(new_data, 'NumEmployees')
print(new_data['NumEmployees'].unique())
```

```python
# Handling Nan values in dataset (it was done manually as answers that we put instead of Nan values differed) (part 1)

#####################################################################
print(new_data.IfRevealedNegativeImpact.value_counts())
print(f"Count of NAN values: {new_data.IfRevealedNegativeImpact.isna().sum()}")

new_data.IfRevealedNegativeImpact.fillna('Prefer not to say', inplace=True)
print(new_data.IfRevealedNegativeImpact.value_counts())

#####################################################################
print(new_data.MentalHealthNegativeProductivity.value_counts())
print(f"Count of NAN values: {new_data.MentalHealthNegativeProductivity.isna().sum()}")

new_data.MentalHealthNegativeProductivity.fillna('Prefer not to say', inplace=True)
print(new_data.MentalHealthNegativeProductivity.value_counts())

#####################################################################
print(new_data.TechCompany.value_counts())
print(f"Count of NAN values: {new_data.TechCompany.isna().sum()}")

new_data.TechCompany.fillna(0.5, inplace=True)
print(new_data.TechCompany.value_counts())

#####################################################################
print(new_data.UnsupportiveOrBadlyHandledResponseToMentalHealthIssueCompany.value_counts())
print(f"Count of NAN values: {new_data.UnsupportiveOrBadlyHandledResponseToMentalHealthIssueCompany.isna().sum()}")

new_data.UnsupportiveOrBadlyHandledResponseToMentalHealthIssueCompany.fillna("I don't know", inplace=True)
print(new_data.UnsupportiveOrBadlyHandledResponseToMentalHealthIssueCompany.value_counts())

#####################################################################
print(new_data.MentalHealthMedicalLeave.value_counts())
print(f"Count of NAN values: {new_data.MentalHealthMedicalLeave.isna().sum()}")

new_data.MentalHealthMedicalLeave.fillna("I don't know", inplace=True)
print(new_data.MentalHealthMedicalLeave.value_counts())

#####################################################################
print(new_data.AgeGroup.value_counts())
print(f"Count of NAN values: {new_data.AgeGroup.isna().sum()}")

new_data.AgeGroup = new_data.AgeGroup.astype('category')
new_data.AgeGroup = new_data.AgeGroup.cat.add_categories(["Prefer not to say"])
new_data.AgeGroup.fillna("Prefer not to say", inplace=True)

print(new_data.AgeGroup.value_counts())
```

```python
# Handling with Nan and ambiguous values (part 2)

#MentalHealthBenefits
def clean_yes_no(data, column_name):
    def process_value(value):
        if pd.isnull(value) or value == '':
            return 'No'
        elif str(value).lower() in ['yes', 'no']:
            return str(value).lower()
        else:
            return "I don't know"

    data[column_name] = data[column_name].apply(process_value)

clean_yes_no(new_data, 'MentalHealthBenefits')

clean_yes_no(new_data, 'EmployerDiscussedMentalHealth')
clean_yes_no(new_data, 'ResourcesForMentalHealth')

#AnonymityProtected
def replace_empty_with_unknown(data, column_name):
    data[column_name] = data[column_name].apply(lambda x: "I don't know" if pd.isnull(x) or x.strip() == '' else x)

replace_empty_with_unknown(new_data, 'AnonymityProtected')

#DiscussMentalHealthWithSupervisor
def clean_discuss_mental_health(value):
    if pd.isnull(value):
        return 'Maybe'
    elif value.lower() == 'yes':
        return 'Yes'
    elif value.lower() == 'no':
        return 'No'
    else:
        return 'Maybe'

new_data['DiscussMentalHealthWithSupervisor'] = new_data['DiscussMentalHealthWithSupervisor'].apply(clean_discuss_mental_health)

#EmployerTakeMentalHealthSeriously
def replace_mental_health_seriously(value):
    if value == 'Yes':
        return 'Yes'
    elif value == 'No':
        return 'No'
    else:
        return "I don't know"

new_data['EmployerTakeMentalHealthSeriously'] = new_data['EmployerTakeMentalHealthSeriously'].apply(replace_mental_health_seriously)
```

```python
# Handling with lower-case and upper-case

#####################################################################
print(new_data.MentalHealthBenefits.value_counts())
print(f"Count of NAN values: {new_data.MentalHealthBenefits.isna().sum()}")

new_data.MentalHealthBenefits.replace(['no', 'No'], 'No', inplace=True)
print(new_data.MentalHealthBenefits.value_counts())

#####################################################################
print(new_data.EmployerDiscussedMentalHealth.value_counts())
print(f"Count of NAN values: {new_data.EmployerDiscussedMentalHealth.isna().sum()}")

new_data.EmployerDiscussedMentalHealth.replace(['no', 'No'], 'No', inplace=True)
print(new_data.EmployerDiscussedMentalHealth.value_counts())

#####################################################################
print(new_data.ResourcesForMentalHealth.value_counts())
print(f"Count of NAN values: {new_data.ResourcesForMentalHealth.isna().sum()}")

new_data.ResourcesForMentalHealth.replace(['no', 'No'], 'No', inplace=True)
print(new_data.ResourcesForMentalHealth.value_counts())
```
